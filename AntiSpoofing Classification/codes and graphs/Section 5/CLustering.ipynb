```python
# Function to convert row name based on condition
def convert_name(row):
    if 'real' in row.lower():
        return 'real'
    elif 'fake' in row.lower():
        return 'fake'
    else:
        return row

# Apply the conversion function to the 'filename' column
labels['label'] = labels['label'].apply(convert_name)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load data
data = features
#labels = pd.read_csv('labels.csv')

# Perform clustering with different number of clusters
clusters = [2, 3, 6, 9, 50]
fig, axs = plt.subplots(1, len(clusters), figsize=(20, 5))

for i, k in enumerate(clusters):
    # Fit KMeans model
    kmeans = KMeans(n_clusters=k, random_state=42).fit(data)

    # Plot results
    axs[i].scatter(data.iloc[:, 0], data.iloc[:, 1], c=kmeans.labels_)
    axs[i].set_title(f'{k} clusters')

plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# Load the data into a DataFrame
data = features

# Split the data into train and test sets
X_train, X_test = train_test_split(data, test_size=0.3, random_state=42)

# Perform KMeans clustering on the training data
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_train)

# Perform Expectation-Maximization (EM) algorithm for classification on the training data
em = GaussianMixture(n_components=2, random_state=42)
em.fit(X_train)

# Predict clusters and class labels for the test data
kmeans_clusters = kmeans.predict(X_test)
em_clusters = em.predict(X_test)

# Plotting the results
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.scatter(X_test['feature 0'], X_test['feature 1'], c=kmeans_clusters)
plt.title('KMeans Clustering')
plt.xlabel('f1')
plt.ylabel('f2')

plt.subplot(1, 2, 2)
plt.scatter(X_test['feature 0'], X_test['feature 1'], c=em_clusters)
plt.title('Expectation-Maximization (EM) Classification')
plt.xlabel('f1')
plt.ylabel('f2')

plt.tight_layout()
plt.show()
```
