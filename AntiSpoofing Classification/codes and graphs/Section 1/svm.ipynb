```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import confusion_matrix,precision_score, accuracy_score, f1_score
dataset=pd.read_csv("/content/drive/MyDrive/dataset.csv")
"""print (dataset.shape)
dataset=dataset.sample(frac=1, replace=False)#shuffling dataset
print (dataset.shape)"""
labels=pd.DataFrame(dataset["label"])
dataset=dataset.drop("label",axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset, labels, test_size=0.3, random_state=42)
```

```python
import numpy as np
svm = SVC()

param_grid = {'C': [0.01, 1, 10], 'kernel': ['linear']}


"""
what thas that above ? check that


"""

grid_search = GridSearchCV(svm, param_grid, cv=5)

grid_search.fit(X_train, y_train['label'].ravel())


# Print the best hyperparameters and corresponding score
print("Best hyperparameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
kernel=grid_search.best_score_
c=grid_search.best_params_
best_params = grid_search.best_params_
```

```python

X_train, X_test, y_train, y_test = train_test_split(dataset, labels, test_size=0.3, random_state=42)
#deg 5
classifier_svm_poly = SVC(kernel="rbf",C=23)#,degree=5)

# Fit the new classifier on the entire training set
classifier_svm_poly.fit(X_train, y_train['label'].ravel())
y_pred = classifier_svm_poly.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')

accuracy = accuracy_score(y_test, y_pred)

f1 = f1_score(y_test, y_pred, average='weighted')

print("Precision: ", precision)
print("Accuracy: ", accuracy)
print("F1 Score: ", f1)
print (cm)
```
