```python
import cv2
import numpy as np
import pandas as pd
import os
csv_train=["x1,x2,y\n"]
csv_test=["x1,x2,y\n"]
```

```python
train_files=os.listdir("cat/data/train/")
dir1="cat/data/train/"
test_files=os.listdir("cat/data/test/")
dir2="cat/data/test/"
```

```python
for j in train_files:
    try:
        img=cv2.imread(dir1+j,cv2.IMREAD_GRAYSCALE)
        
        for i in range(64):
            for f in range(64):
                csv_train.append(str(i)+","+str(f)+","+str(img[i][f])+"\n")
    except : pass
for j in test_files:
    try :
        img=cv2.imread(dir2+j,cv2.IMREAD_GRAYSCALE)
        for i in range(64):
            for f in range(64):
                csv_test.append(str(i)+","+str(f)+","+str(img[i][f])+"\n") 
    except : pass
```

```python
with open ("test_cat.csv","w") as f:
    f.writelines(csv_test[:4000000])
```

```python
with open ("train_cat.csv","w") as f:
    f.writelines(csv_train[:4096000])
```
